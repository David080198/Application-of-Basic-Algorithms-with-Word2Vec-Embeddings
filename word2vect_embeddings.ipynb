```python
import pandas as pd

ruta = "F:/nuevo_conocimiento/CODIGO DE TESIS (MUY IMPORTANTE)/Experimentos_con_modelos_basicos/dataset_preprocesado/dataset_abstract_preprocesado.csv"

datos = pd.read_csv(ruta)
datos
```

```python
from gensim.models import Word2Vec
import numpy as np
from tqdm import tqdm
```

```python
lista_de_listas = datos['abstract_preprocesado'].tolist()
# Definir la longitud de los vectores
longitud_vector = 10000

# Entrenar el modelo Word2Vec
modelo = Word2Vec(lista_de_listas, vector_size=longitud_vector, min_count=1)
```

```python
def calcular_vec2vect(texto):
    vectores_palabras = [modelo.wv[palabra] for palabra in texto if palabra in modelo.wv]
    if vectores_palabras:
        promedio_vector = np.mean(vectores_palabras, axis=0)
        #print(promedio_vector)
    return promedio_vector
```

```python

vectores_word2vect = []
for texto in tqdm(lista_de_listas,desc="Procesando"):
    vectores_word2vect.append(calcular_vec2vect(texto))
```

```python
word2vect = pd.DataFrame(vectores_word2vect)
etiquetas = datos['cat_one_hot']
df_junto = pd.concat([word2vect, etiquetas], axis=1)
df_junto
```
